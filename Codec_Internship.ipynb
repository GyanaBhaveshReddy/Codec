```python
class user:

    def __init__(self, userid, pin, balance):
        self.userid = userid.lower().strip()
        self.pin = pin.strip()
        self.balance = balance
        self.history = []

    def deposit(self, amount):
        self.balance += amount
        self.history.append(f"Deposited ₹{amount:.2f}")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.history.append(f"Withdrawn ₹{amount:.2f}")
            return True
        return False

    def transfer(self, receiver, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            receiver.balance += amount
            self.history.append(f"Transferred ₹{amount:.2f} to {receiver.userid}")
            receiver.history.append(f"Received ₹{amount:.2f} from {self.userid}")
            return True
        return False

    def showhistory(self):
        if not self.history:
            print("\nNo transactions yet")
        else:
            print("\n--- Transaction History ---")
            for t in self.history:
                print(t)


class atm:

    def __init__(self):
        self.users = {}
        self.current = None

    def start(self):
        print("===== Welcome to ATM Interface =====")

        while True:
            print("\n1. Create Account")
            print("2. Login")
            print("3. Exit")

            choice = input("Choose option: ")

            if choice == "1":
                self.create()
            elif choice == "2":
                self.login()
            elif choice == "3":
                print("Thank you for visiting")
                break
            else:
                print("Invalid choice")

    def create(self):
        userid = input("Enter User ID: ").strip().lower()

        if userid in self.users:
            print("User already exists")
            return

        pin = input("Set 4-digit PIN: ").strip()

        if len(pin) != 4 or not pin.isdigit():
            print("Invalid PIN")
            return

        try:
            balance = float(input("Enter initial deposit: "))
        except:
            print("Invalid amount")
            return

        self.users[userid] = user(userid, pin, balance)
        print("Account created successfully")

    def login(self):
        userid = input("Enter User ID: ").strip().lower()
        pin = input("Enter PIN: ").strip()

        if userid in self.users and self.users[userid].pin == pin:
            self.current = self.users[userid]
            print(f"Welcome {self.current.userid}")
            self.menu()
        else:
            print("Invalid credentials")

    def menu(self):
        while True:
            print("\n===== ATM MENU =====")
            print("1. Transaction History")
            print("2. Withdraw")
            print("3. Deposit")
            print("4. Transfer")
            print("5. Check Balance")
            print("6. Logout")

            choice = input("Enter choice: ")

            if choice == "1":
                self.current.showhistory()

            elif choice == "2":
                try:
                    amt = float(input("Enter amount: "))
                    if self.current.withdraw(amt):
                        print(f"Withdraw successful ₹{self.current.balance:.2f}")
                    else:
                        print("Failed transaction")
                except:
                    print("Invalid input")

            elif choice == "3":
                try:
                    amt = float(input("Enter amount: "))
                    self.current.deposit(amt)
                    print(f"Deposit successful ₹{self.current.balance:.2f}")
                except:
                    print("Invalid input")

            elif choice == "4":
                rid = input("Enter receiver ID: ").strip().lower()

                if rid not in self.users:
                    print("User not found")
                    continue

                try:
                    amt = float(input("Enter amount: "))
                    if self.current.transfer(self.users[rid], amt):
                        print(f"Transfer successful ₹{self.current.balance:.2f}")
                    else:
                        print("Transfer failed")
                except:
                    print("Invalid amount")

            elif choice == "5":
                print(f"Balance: ₹{self.current.balance:.2f}")

            elif choice == "6":
                print("Logged out")
                self.current = None
                break

            else:
                print("Invalid choice")


a = atm()
a.start()
```
